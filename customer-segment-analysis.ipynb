```python
import pandas as pd # for data manupulation
import numpy as np # to work with array
import matplotlib.pyplot as plt # for plotting
```

```python
df = pd.read_csv('travel_tide.csv') # load dataset
```

```python
df.columns
```

# Free Hotel Meal Perk

```python
df['total_hotel_spent'] = df['rooms'] * df['hotel_per_room_usd'] * df['nights'] # total spend on hotel
```

```python
df
```

```python
new_df = df[['user_id', 'total_hotel_spent']] # get the related column
new_df
```

```python
# get total spending on hotel by each user, sort it descending order on total spent
new_df = new_df.groupby('user_id', as_index=False).sum('total_hotel_spent').sort_values('total_hotel_spent', ascending=False)
```

```python
new_df
```

```python
new_df['total_hotel_spent'].describe() # summery
```

```python
new_df['total_hotel_spent'].hist() # plot histogram
```

```python
plt.boxplot(new_df['total_hotel_spent']) # plot boxplot
```

```python
# Calculate Q1, Q3, and IQR
Q1 = new_df['total_hotel_spent'].quantile(0.25)
Q3 = new_df['total_hotel_spent'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_filtered = new_df[(new_df['total_hotel_spent'] >= lower_bound) & (new_df['total_hotel_spent'] <= upper_bound)]

print("Original DataFrame:")
print(new_df)
print("\nDataFrame after removing outliers:")
print(df_filtered)
```

```python
df_filtered['total_hotel_spent'].hist()
```

```python
plt.boxplot(df_filtered['total_hotel_spent'])
```

```python
df_filtered['total_hotel_spent'].describe()
```

```python
# Create conditional DataFrame column by np.select function.
# giving free meal perks
conditions = [new_df['total_hotel_spent'] > 2293]

choices = ['yes']
new_df['free_meal'] = np.select(conditions, choices, default='no')
print(new_df)
```

```python
new_df[new_df['free_meal'] == 'yes'].count() # total users that got the perk
```

```python
segmented_df = new_df[['user_id', 'free_meal']] # new dataframe for segmented dataset
segmented_df
```

# Free Checked Bag

```python
new_df = df[['user_id', 'checked_bags']] # getting the related column
```

```python
# Calculate total checked bags for each user
# bag_counts = df.groupby('user_id')['checked_bags'].sum()

new_df = new_df.groupby('user_id', as_index=False).sum('checked_bags').sort_values('checked_bags', ascending=False)
```

```python
new_df.checked_bags.describe()
```

```python
plt.boxplot(new_df.checked_bags)
```

```python
# Calculate Q1, Q3, and IQR
Q1 = new_df['checked_bags'].quantile(0.25)
Q3 = new_df['checked_bags'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_filtered = new_df[(new_df['checked_bags'] >= lower_bound) & (new_df['checked_bags'] <= upper_bound)]

print("Original DataFrame:")
print(new_df)
print("\nDataFrame after removing outliers:")
print(df_filtered)
```

```python
plt.boxplot(df_filtered.checked_bags)
```

```python
df_filtered.checked_bags.describe()
```

```python
new_df[new_df['checked_bags'] > 2].count()
```

```python
# Create conditional DataFrame column by np.select function.
# giving the perk who has more than 2 checked bags
conditions = [new_df['checked_bags'] > 2]

choices = ['yes']
new_df['free_checked_bags'] = np.select(conditions, choices, default='no')
print(new_df)
```

```python
# merging to main segmented dataset
segmented_df = pd.merge(segmented_df, new_df[['user_id', 'free_checked_bags']], on='user_id', how='inner')
segmented_df
```

```python
new_df
```

```python
new_df['no_cancelled_trip'].describe()
```

```python
new_df[new_df['no_cancelled_trip'] > 4].count()
```

```python
# Create conditional DataFrame column by np.select function.
# giving the perk who has more than 4 trip
conditions = [new_df['trip_id'] > 4]

choices = ['yes']
new_df['no_cancellation_fees'] = np.select(conditions, choices, default='no')
print(new_df)
```

```python
# merging
segmented_df = pd.merge(segmented_df, new_df[['user_id', 'no_cancellation_fees']], on='user_id', how='inner')
segmented_df
```

# Exclusive Discounts

```python
# all the trip with flight and hotel booking and have flight or hotel discount with no cancellation
new_df = df[
    ((df['trip_id.1'].notnull()) |
    (df['trip_id.2'].notnull())) &
    (df['cancellation'] == False) &
    ((df['flight_discount'] == True) | (df['hotel_discount'] == True))
]
new_df
```

```python
new_df = new_df[['user_id']]
new_df
```

```python
# geting the total discounted trip by each user
new_df = new_df.groupby('user_id', as_index=False).value_counts().sort_values('count', ascending=False)
new_df['discounted_trip'] = new_df['count']
new_df
```

```python
new_df['discounted_trip'].describe()
```

```python
plt.boxplot(new_df['discounted_trip'])
```

```python
new_df[new_df['discounted_trip'] > 2].count()
```

```python
# Create conditional DataFrame column by np.select function.
# giving the perk who has more than 2 trip
conditions = [new_df['discounted_trip'] > 2]

choices = ['yes']
new_df['exclusive_discounts'] = np.select(conditions, choices, default='no')
print(new_df)
```

```python
# merging
segmented_df = pd.merge(segmented_df, new_df[['user_id', 'exclusive_discounts']], on='user_id', how='left')
segmented_df
```

```python
segmented_df.isna().sum() # calculating the total NA cell
```

# 1 night free hotel with flight

```python
# all the trip with flight and hotel booking
new_df = df[(df['trip_id.1'].notnull()) & (df['trip_id.2'].notnull())]

new_df
```

```python
# booked return flight
new_df = new_df[new_df['return_flight_booked'] == True]
new_df
```

```python
new_df = new_df[['user_id']]
new_df
```

```python
# counting the total trip by each user
new_df = new_df.groupby('user_id', as_index=False).value_counts().sort_values('count', ascending=False)
new_df['trip_with_return_flight'] = new_df['count']
new_df
```

```python
new_df['trip_with_return_flight'].describe()
```

```python
new_df[new_df['trip_with_return_flight'] > 3].count()
```

```python
# Create conditional DataFrame column by np.select function.
conditions = [new_df['trip_with_return_flight'] > 3]
# giving the perk who trip at least 3 times
choices = ['yes']
new_df['free_hotel_night'] = np.select(conditions, choices, default='no')
print(new_df)
```

```python
# merging
segmented_df = pd.merge(segmented_df, new_df[['user_id', 'free_hotel_night']], on='user_id', how='left')
segmented_df
```

```python
# filling all the NA
segmented_df['exclusive_discounts'] = segmented_df['exclusive_discounts'].fillna('no')
segmented_df['free_hotel_night'] = segmented_df['free_hotel_night'].fillna('no')
segmented_df.isna().sum()
```

```python
# save the dataset
segmented_df.to_csv('travelTide_customer_segmented_dataset.csv', index=False)
```
