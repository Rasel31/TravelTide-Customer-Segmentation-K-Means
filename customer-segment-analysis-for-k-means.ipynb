```python
import pandas as pd # for data manupulation
import matplotlib.pyplot as plt # for plotting
```

```python
df = pd.read_csv('travel_tide.csv') # load dataset
```

```python
df
```

```python
df.columns
```

# Calculate total hotel spent without discount

```python
df['total_hotel_spent'] = df['rooms'] * df['hotel_per_room_usd'] * df['nights'] # total spend on hotel
```

```python
df
```

```python
new_df = df[['user_id', 'total_hotel_spent']] # get the related column
new_df
```

```python
# get total spending on hotel by each user, sort it descending order on total spent
new_df = new_df.groupby('user_id', as_index=False).sum('total_hotel_spent').sort_values('total_hotel_spent', ascending=False)
```

```python
new_df
```

```python
segmented_kmeans = new_df[['user_id', 'total_hotel_spent']] # making new datafram for clustering
segmented_kmeans
```

# Calculate total checked bags

```python
new_df = df[['user_id', 'checked_bags']] # getting the related column
```

```python
# Calculate total checked bags for each user
new_df = new_df.groupby('user_id', as_index=False).sum('checked_bags').sort_values('checked_bags', ascending=False)
```

```python
new_df
```

```python
# merging to clestering dataframe
segmented_kmeans = pd.merge(segmented_kmeans, new_df[['user_id', 'checked_bags']], on='user_id', how='inner')
segmented_kmeans
```

# Calculate total no canceled trip by each user

```python
# exclude cancel trip
no_cancel = df[df['cancellation'] != True]
no_cancel
```

```python
new_df = no_cancel[['user_id', 'trip_id']] # getting the related column
```

```python
# get the total trip by each user
new_df = new_df.groupby('user_id', as_index=False)['trip_id'].count().sort_values('trip_id', ascending=False)
```

```python
new_df
```

```python
new_df['no_cancelled_trip'] = new_df['trip_id'] # converting to a meaningful column
new_df
```

```python
# merging
segmented_kmeans = pd.merge(segmented_kmeans, new_df[['user_id', 'no_cancelled_trip']], on='user_id', how='inner')
segmented_kmeans
```

# Calculate discounted trip by user

```python
# all the trip with flight and hotel booking having flight or hotel discount with no cancellation
new_df = df[
    ((df['trip_id.1'].notnull()) |
    (df['trip_id.2'].notnull())) &
    (df['cancellation'] == False) &
    ((df['flight_discount'] == True) | (df['hotel_discount'] == True))
]
new_df
```

```python
new_df = new_df[['user_id']]
new_df
```

```python
# geting the total discounted trip by each user
new_df = new_df.groupby('user_id', as_index=False).value_counts().sort_values('count', ascending=False)
new_df['discounted_trip'] = new_df['count']
new_df
```

```python
# merging
segmented_kmeans = pd.merge(segmented_kmeans, new_df[['user_id', 'discounted_trip']], on='user_id', how='left')
segmented_kmeans
```

# calculate total trip with return flight

```python
# all the trip with flight and hotel booking with no cancellation
new_df = df[(df['trip_id.1'].notnull()) & (df['trip_id.2'].notnull()) & (df['cancellation'] == False)]
new_df
```

```python
# booked return flight
new_df = new_df[new_df['return_flight_booked'] == True]
new_df
```

```python
new_df = new_df[['user_id']]
new_df
```

```python
# counting the total trip by each user
new_df = new_df.groupby('user_id', as_index=False).value_counts().sort_values('count', ascending=False)
new_df['trip_with_return_flight'] = new_df['count']
new_df
```

```python
# merging
segmented_kmeans = pd.merge(segmented_kmeans, new_df[['user_id', 'trip_with_return_flight']], on='user_id', how='left')
segmented_kmeans
```

```python
segmented_kmeans.isna().sum() # summing up total NA cell
```

```python
# filling NA cell with 0.0
segmented_kmeans['trip_with_return_flight'] = segmented_kmeans['trip_with_return_flight'].fillna(0.0)
segmented_kmeans['discounted_trip'] = segmented_kmeans['discounted_trip'].fillna(0.0)
segmented_kmeans.isna().sum()
```

```python
# saving the dataset for kmeans clustering
segmented_kmeans.to_csv('travelTide_customer_segmented_dataset-kmeans.csv', index=False)
```
